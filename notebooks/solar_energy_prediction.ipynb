```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/solar_clean.csv")

# Display first rows
df.head()

# Plot correlation
plt.scatter(df["Temperature"], df["GHI"], color='orange')
plt.title("Temperature vs Solar Energy Output (GHI)")
plt.xlabel("Temperature (°C)")
plt.ylabel("GHI (W/m²)")
plt.show()
```
